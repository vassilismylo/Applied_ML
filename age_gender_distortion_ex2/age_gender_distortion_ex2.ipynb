```python
male_labels = [
    "chairman of the board",
    "elected official",
    "director of research",
    "chief of staff",
    "military personnel",
]
```

```python
female_labels = [
    "intern",
    "novice",
    "cook",
    "secretary",
    "homoeopath"
]
```

```python
male_df = gpt_dim[gpt_dim["Social_Category"].isin(male_labels)]

```

```python
female_df = gpt_dim[gpt_dim["Social_Category"].isin(female_labels)]
```

```python
rest_df = gpt_dim[~gpt_dim["Social_Category"].isin(male_labels)]
rest_df = rest_df[~rest_df["Social_Category"].isin(female_labels)]
```

```python
annot = annot.merge(
    gpt_dim[[
        "Social_Category",
        "gender_norm_score",
        "age_norm_score"
    ]],
    left_on="label",
    right_on="Social_Category",
    how="left"
)
```

```python
annot = annot.rename(columns={
    "gender_norm_score": "x",
    "age_norm_score": "y"
})
```

```python
annot
```

Created using plotnine.

Started from a basic scatter + lm fit. Axis limits are fixed and the fit is
extended manually to the full range. Background points are hollow to avoid
overplotting.

Highlighted points are handled separately. Label positions are defined
manually (lx / ly) in a small annotation table, merged back to the data,
and connected to points with line segments. Colors are passed through an
identity scale to avoid ggplot’s default remapping.


```python
p = (
    ggplot(gpt_dim, aes(x="gender_norm_score", y="age_norm_score"))

    # background cloud
   + geom_point(
    data=rest_df,
    mapping=aes(x="gender_norm_score", y="age_norm_score"),
    size=0.8,
    fill="none",
    color="black",
    stroke=0.6,
    shape="o"
)


    # regression
    + geom_smooth(method="lm", se=False, color="red", fullrange=True)

    + geom_point(
    data=annot,
    mapping=aes(
        x="x",
        y="y",
        fill="color"  # column name
    ),
    # fill="color",
    size=3.5,
    color="black",
    stroke=0.6,
    shape="o"
)



    # connector lines
    + geom_segment(
        data=annot,
        mapping=aes(
            x="x", y="y",
            xend="lx", yend="ly"
        ),
        size=0.2,
        color="black"
    )

    # labels (ALL centered, as you want)
    + geom_text(
        data=annot,
        mapping=aes(
            x="lx", y="ly",
            label="label"
        ),
        ha="center",
        va="center",
        size=10
    )
    + scale_fill_identity()

    # axes
    + labs(
        x="Gender Association\n(Female–Male Dimension)",
        y="Age Association\n(Young–Old Dimension)"
    )
    + scale_x_continuous(limits=(-0.25, 1.35), expand=(0, 0))
    + scale_y_continuous(limits=(-0.20, 1.20), expand=(0, 0))
)
p.save(
    "age_gender_plot.svg",
    width=10,
    height=8,
    units="in"
)

```

lets make this interactive

```python
import altair as alt
```

```python
base = alt.Chart(rest_df).mark_point(
    fill='lightgray',
    stroke='black',
    strokeWidth=0.1,
    size=40
).encode(
    x=alt.X(
        'gender_norm_score:Q',
        title='Gender Association (Female–Male Dimension)'
    ),
    y=alt.Y(
        'age_norm_score:Q',
        title='Age Association (Young–Old Dimension)'
    ),
    tooltip=[
        'Social_Category:N',
        alt.Tooltip('gender_norm_score:Q', format='.3f'),
        alt.Tooltip('age_norm_score:Q', format='.3f')
    ]
)
reg_line = alt.Chart(gpt_dim).transform_regression(
    'gender_norm_score',
    'age_norm_score'
).mark_line(
    color='red',
    strokeWidth=2
).encode(
    x='gender_norm_score:Q',
    y='age_norm_score:Q'
)
highlight_pts = alt.Chart(annot).mark_point(
    filled=True,
    stroke='black',
    strokeWidth=0.2,
    size=120
).encode(
    x='x:Q',
    y='y:Q',
    fill=alt.Fill('color:N', scale=None),
    tooltip=['label:N', 'x:Q', 'y:Q']
)

hover = alt.selection_point(
    fields=['label'],
    on='mouseover',
    empty=False
)
highlight_pts = highlight_pts.add_params(hover).encode(
    opacity=alt.condition(hover, alt.value(1), alt.value(0.7))
)

chart = (
    base
    + reg_line
    + highlight_pts
).properties(
    width=700,
    height=550
)
chart
```
