The table in the assignment uses two level column headers so i need to do this in my data frame. Here i got help from chatgpt. As i understand it, i replace normal columns names with tuples. Each column is a tuple, with the first value the 1st level header and the second the 2nd level header.

```python
social_table2 = social_table.copy()

social_table2.columns = [
    ("dimension", ""),
    ("indicator", ""),
    ("date", "first"),
    ("date", "last"),
    ("value", "first"),
    ("value", "last"),
]

social_table2.columns = pd.MultiIndex.from_tuples(social_table2.columns)

social_table2
```

Lets make the dimentions capitalized

```python
social_table2["dimension"] = social_table2["dimension"].str.title()
social_table2
```

From <https://pandas.pydata.org/docs/user_guide/style.html> (pandas Table Visualization documentation), i found how to set the css properties of the table so that it matches the required one: There were 3 problems that the regular to_html() or style.to_html() tables were missing:

1) The column headers were not centered
2) the first 2 column data were not left aligned
3) the values were not up to the first 2 digits
I correct all of these here:

```python
from pandas import IndexSlice as idx

styled = (
    social_table2.style
    # remove the index
    .hide(axis="index")
    # center the column headers 
    .set_table_styles([
        {"selector": "th.col_heading",        "props": [("text-align", "center")]},
        {"selector": "th.col_heading.level0", "props": [("text-align", "center")]},
        {"selector": "th.col_heading.level1", "props": [("text-align", "center")]},
    ])
    # left-align first two columns (dimension, indicator)
    .set_properties(
        subset=idx[:, [("dimension", ""), ("indicator", "")]],
        **{"text-align": "left"}
    )
     # control how numbers are displayed
    .format({
        ("value", "first"): "{:.2f}",
        ("value", "last"):  "{:.2f}",
    })
)
styled
```

I save the html code that i generated into a file (social_table.md)

```python
social_table = styled.to_html()
with open("social_table.md", "w", encoding="utf-8") as f:
    f.write(social_table)
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# ---------------------------------------------------------
# 1. Load data
# ---------------------------------------------------------
df = pd.read_csv("full_data.csv")

# --- social ---
soc = df[df["domain"] == "social"].copy()
# clip negatives, and keep a copy for colour scaling
soc["shortfall_clip"] = soc["shortfallOvershoot_pct"].clip(lower=0)
soc["shortfall_frac"] = (soc["shortfallOvershoot_pct"] / 100).clip(lower=0)
dimensions = list(dict.fromkeys(soc["dimension"]))

# --- ecological ---
eco = df[df["domain"] == "ecological"].copy()
eco_dims = list(dict.fromkeys(eco["dimension"]))

# overshoot as a fraction, scaled by a FIXED constant (not the max)
eco["overshoot_raw"] = eco["shortfallOvershoot_pct"].clip(lower=0)
ECO_SCALE = 200.0   # bigger -> shorter spikes overall, smaller -> longer
eco["overshoot_frac"] = eco["overshoot_raw"] / ECO_SCALE   # NO upper clip

# ---------------------------------------------------------
# Colour scaling (continuous) – separate for social & ecological
# ---------------------------------------------------------
# social range
SOC_MIN = soc["shortfall_clip"].min()
SOC_MAX = soc["shortfall_clip"].max()
if SOC_MAX <= SOC_MIN:
    SOC_MAX = SOC_MIN + 1.0  # avoid divide-by-zero

# ecological range (use overshoot_raw)
ECO_MIN = eco["overshoot_raw"].min()
ECO_MAX = eco["overshoot_raw"].max()
if ECO_MAX <= ECO_MIN:
    ECO_MAX = ECO_MIN + 1.0

# RGB for light and dark red (from your hexes)
LIGHT_RGB = np.array([0xFD, 0x8D, 0xA9])  # #FD8DA9
DARK_RGB  = np.array([0x50, 0x09, 0x24])  # #500924

def _interp_red(t: float) -> str:
    """
    t in [0,1] -> hex colour between LIGHT_RGB and DARK_RGB.
    0 = light, 1 = dark.
    """
    t = max(0.0, min(1.0, float(t)))
    rgb = (1 - t) * LIGHT_RGB + t * DARK_RGB
    r, g, b = rgb.astype(int)
    return f"#{r:02X}{g:02X}{b:02X}"

def get_social_color(pct: float) -> str:
    """
    Continuous colour for social bar based on its shortfall %.
    """
    val = max(0.0, pct)
    t = (val - SOC_MIN) / (SOC_MAX - SOC_MIN)
    return _interp_red(t)

def get_eco_color(overshoot_raw: float) -> str:
    """
    Continuous colour for ecological bar based on overshoot_raw.
    """
    val = max(0.0, overshoot_raw)
    t = (val - ECO_MIN) / (300 - ECO_MIN)
    return _interp_red(t)

# ---------------------------------------------------------
# 2. Build flat indicator list so each social indicator has same width
# ---------------------------------------------------------
indicator_pairs = []          # social [(dim, ind), ...]
dim_first_idx = {}
dim_counts = {d: 0 for d in dimensions}

idx = 0
for dim in dimensions:
    dim_data = soc[soc["dimension"] == dim]
    inds = list(dict.fromkeys(dim_data["indicator"]))
    for ind in inds:
        indicator_pairs.append((dim, ind))
        if dim not in dim_first_idx:
            dim_first_idx[dim] = idx
        dim_counts[dim] += 1
        idx += 1

n_indicators = len(indicator_pairs)
ind_span = 360 / n_indicators

# ---------------------------------------------------------
# 2b. Build eco indicator list (equal-width wedges by INDICATOR)
# ---------------------------------------------------------
eco_indicator_pairs = []       # ecological [(dim, ind), ...]
eco_dim_first_idx = {}
eco_dim_counts = {d: 0 for d in eco_dims}

idx = 0
for dim in eco_dims:
    dim_data = eco[eco["dimension"] == dim]
    inds = list(dict.fromkeys(dim_data["indicator"]))
    for ind in inds:
        eco_indicator_pairs.append((dim, ind))
        if dim not in eco_dim_first_idx:
            eco_dim_first_idx[dim] = idx
        eco_dim_counts[dim] += 1
        idx += 1

n_eco_indicators = len(eco_indicator_pairs)
eco_span = 360 / n_eco_indicators

# ---------------------------------------------------------
# 3. Geometry
# ---------------------------------------------------------
R_RING_IN  = 1.0     # outer edge of social wedges
R_BAR_MIN  = 0.0     # 100% shortfall goes to center

# Area scaling for social bars
K = R_RING_IN**2 - R_BAR_MIN**2

# --- radii for the 3 green rings ---
R_SF_IN  = R_RING_IN          # inner dark ring (Social foundation)
R_SF_OUT = R_SF_IN + 0.06

R_DIM_IN  = R_SF_OUT          # middle light-green ring (social dimension labels)
R_DIM_OUT = R_DIM_IN + 0.20

R_ECO_IN  = R_DIM_OUT         # outer dark ring (Ecological ceiling)
R_ECO_OUT = R_ECO_IN + 0.06

# --- ecological wedge geometry (outside the dark ring) ---
R_ECO_BAR_IN  = R_ECO_OUT             # inner radius of eco bars
R_ECO_BAR_MAX = R_ECO_BAR_IN + 1.0    # reference outer radius for overshoot_frac = 1
K_ECO = R_ECO_BAR_MAX**2 - R_ECO_BAR_IN**2

R_RING_OUT = R_ECO_BAR_MAX            # overall outer radius of whole figure

# colors
COLOR_DARK_GREEN  = "rgba(0, 90, 40, 1.0)"
COLOR_LIGHT_GREEN = "rgba(90, 170, 70, 1.0)"

# ---------------------------------------------------------
# 4. Outline wedge (full social indicator) – annular sector
# ---------------------------------------------------------
def add_inward_outline(fig, theta_start, theta_end):
    theta_outer = np.linspace(theta_start, theta_end, 40)
    theta_inner = theta_outer[::-1]

    t_out = np.radians(theta_outer)
    t_in  = np.radians(theta_inner)

    r_out = np.full_like(theta_outer, R_RING_IN)
    r_in  = np.full_like(theta_inner, R_BAR_MIN)

    x_out = r_out * np.cos(t_out)
    y_out = r_out * np.sin(t_out)
    x_in  = r_in * np.cos(t_in)
    y_in  = r_in * np.sin(t_in)

    fig.add_trace(go.Scatter(
        x=np.concatenate([x_out, x_in]),
        y=np.concatenate([y_out, y_in]),
        fill="toself",
        fillcolor="rgba(200, 30, 90, 0.10)",   # very light pink
        line=dict(width=0),
        hoverinfo="skip",
        showlegend=False
    ))

# ---------------------------------------------------------
# 5. Inward area-scaled social bar – annular sector + invisible hover marker
# ---------------------------------------------------------
def add_inward_bar(fig, frac, theta_start, theta_end, hover_text, color):
    if frac <= 0:
        return

    # inner radius from area scaling
    r_bar_in = np.sqrt(R_RING_IN**2 - K * frac)

    theta_outer = np.linspace(theta_start, theta_end, 40)
    theta_inner = theta_outer[::-1]
    t_out = np.radians(theta_outer)
    t_in  = np.radians(theta_inner)

    r_out = np.full_like(theta_outer, R_RING_IN)
    r_in  = np.full_like(theta_inner, r_bar_in)

    x_out = r_out * np.cos(t_out)
    y_out = r_out * np.sin(t_out)
    x_in  = r_in * np.cos(t_in)
    y_in  = r_in * np.sin(t_in)

    xs = np.concatenate([x_out, x_in])
    ys = np.concatenate([y_out, y_in])

    # 1) actual colored bar (no direct hover)
    fig.add_trace(go.Scatter(
        x=xs,
        y=ys,
        mode="lines",
        fill="toself",
        fillcolor=color,
        line=dict(width=0),
        hoverinfo="skip",
        showlegend=False
    ))

    # 2) large invisible markers along the radius for hover
    mid_ang = 0.5 * (theta_start + theta_end)
    mid_t = np.radians(mid_ang)

    # three radii from inner to outer edge of the bar
    radii = np.linspace(r_bar_in, R_RING_IN, 3)

    x_mark = radii * np.cos(mid_t)
    y_mark = radii * np.sin(mid_t)

    fig.add_trace(go.Scatter(
        x=x_mark,
        y=y_mark,
        mode="markers",
        marker=dict(
            size=40,                   # big hitbox
            color="rgba(0,0,0,0)",     # fully invisible
        ),
        hovertemplate=hover_text + "<extra></extra>",
        showlegend=False
    ))

# ---------------------------------------------------------
# --- ECOLOGICAL: outward bars  ---
# ---------------------------------------------------------
def add_outward_bar(fig, frac, theta_start, theta_end, hover_text, color):
    """Area-scaled eco overshoot bar from R_ECO_BAR_IN outward."""
    if frac <= 0:
        return

    # area scaling with overshoot_frac (can be >1 if very large overshoot)
    r_bar_out = np.sqrt(R_ECO_BAR_IN**2 + K_ECO * frac)

    theta_outer = np.linspace(theta_start, theta_end, 40)
    theta_inner = theta_outer[::-1]
    t_out = np.radians(theta_outer)
    t_in  = np.radians(theta_inner)

    r_out = np.full_like(theta_outer, r_bar_out)
    r_in  = np.full_like(theta_inner, R_ECO_BAR_IN)

    x_out = r_out * np.cos(t_out)
    y_out = r_out * np.sin(t_out)
    x_in  = r_in * np.cos(t_in)
    y_in  = r_in * np.sin(t_in)

    xs = np.concatenate([x_out, x_in])
    ys = np.concatenate([y_out, y_in])

    # 1) coloured overshoot bar (no direct hover)
    fig.add_trace(go.Scatter(
        x=xs,
        y=ys,
        mode="lines",
        fill="toself",
        fillcolor=color,
        line=dict(width=0),
        hoverinfo="skip",
        showlegend=False
    ))

    # 2) large invisible markers along the ecological bar
    mid_ang = 0.5 * (theta_start + theta_end)
    mid_t = np.radians(mid_ang)

    radii = np.linspace(R_ECO_BAR_IN, r_bar_out, 5)

    x_mark = radii * np.cos(mid_t)
    y_mark = radii * np.sin(mid_t)

    fig.add_trace(go.Scatter(
        x=x_mark,
        y=y_mark,
        mode="markers",
        marker=dict(
            size=60,
            color="rgba(0,0,0,0)",
        ),
        hovertemplate=hover_text + "<extra></extra>",
        showlegend=False
    ))

# ---------------------------------------------------------
# 8. Helper to draw a ring between r_in and r_out
# ---------------------------------------------------------
def add_ring(fig, r_in, r_out, color):
    theta = np.linspace(0, 360, 361)
    t = np.radians(theta)
    x_out = r_out * np.cos(t)
    y_out = r_out * np.sin(t)
    x_in  = r_in  * np.cos(t[::-1])
    y_in  = r_in  * np.sin(t[::-1])

    fig.add_trace(go.Scatter(
        x=np.concatenate([x_out, x_in]),
        y=np.concatenate([y_out, y_in]),
        fill="toself",
        fillcolor=color,
        line=dict(width=0),
        hoverinfo="skip",
        showlegend=False
    ))

# ---------------------------------------------------------
# helper to put a WORD along an arc (per letter)
# ---------------------------------------------------------
def add_arc_label(fig, text, radius, angle_center, arc_span_deg, size=12):
    chars = list(text)
    n = len(chars)
    if n == 0:
        return

    angles = np.linspace(
        angle_center - arc_span_deg / 2,
        angle_center + arc_span_deg / 2,
        n
    )

    # reverse angles so letters read left-to-right along the arc
    for ch, ang in zip(chars, angles[::-1]):
        t = np.radians(ang)
        x = radius * np.cos(t)
        y = radius * np.sin(t)

        # same rotation rule as for dimensions
        pos_angle = ((ang + 180) % 360) - 180
        if pos_angle >= 0:
            tang = 90 - pos_angle
        else:
            tang = -90 - pos_angle

        fig.add_annotation(
            x=x,
            y=y,
            text=ch,
            showarrow=False,
            font=dict(size=size, color="white"),
            textangle=tang,
            xanchor="center",
            yanchor="middle"
        )

# ---------------------------------------------------------
# 6. Build figure
# ---------------------------------------------------------
fig = go.Figure()

# 6a. Social wedges (equal-width by indicator)
for k, (dim, ind) in enumerate(indicator_pairs):
    row = soc[(soc["dimension"] == dim) & (soc["indicator"] == ind)].iloc[0]
    frac = row["shortfall_frac"]
    pct  = row["shortfall_clip"]
    color = get_social_color(pct)

    start_ang = k * ind_span
    end_ang   = (k + 1) * ind_span

    # light full outline wedge
    add_inward_outline(fig, start_ang, end_ang)

    # actual bar with hover "indicator: xx.x%"
    hover = f"{ind}: {row['shortfallOvershoot_pct']:.1f}%"
    add_inward_bar(fig, frac, start_ang, end_ang, hover, color)

# ---------------------------------------------------------
# 7. White radial separator lines (up to outer edge of social wedges)
# ---------------------------------------------------------
for k in range(n_indicators):
    ang = k * ind_span
    rad = np.radians(ang)
    x0, y0 = 0.0, 0.0
    x1, y1 = R_RING_IN * np.cos(rad), R_RING_IN * np.sin(rad)

    fig.add_trace(go.Scatter(
        x=[x0, x1],
        y=[y0, y1],
        mode="lines",
        line=dict(color="rgba(255,255,255,0.7)", width=1),
        hoverinfo="skip",
        showlegend=False
    ))

# ---------------------------------------------------------
# 9. Draw the three green rings (social + dimension + eco base)
# ---------------------------------------------------------
add_ring(fig, R_SF_IN,  R_SF_OUT,  COLOR_DARK_GREEN)   # inner dark
add_ring(fig, R_DIM_IN, R_DIM_OUT, COLOR_LIGHT_GREEN)  # middle light (social dims)
add_ring(fig, R_ECO_IN, R_ECO_OUT, COLOR_DARK_GREEN)   # outer dark (eco ceiling)

# ---------------------------------------------------------
# 10. Social dimension labels inside the light-green ring
# ---------------------------------------------------------
r_dim_label = 0.5 * (R_DIM_IN + R_DIM_OUT)

for dim in dimensions:
    first = dim_first_idx[dim]
    count = dim_counts[dim]
    center_index = first + count / 2.0     # middle indicator in this dimension
    ang = center_index * ind_span          # angle in degrees
    t_lab = np.radians(ang)

    x = r_dim_label * np.cos(t_lab)
    y = r_dim_label * np.sin(t_lab)

    # match assignment's rotation rule
    pos_angle = ((ang + 180) % 360) - 180

    if pos_angle >= 0:
        tang = 90 - pos_angle
    else:
        tang = -90 - pos_angle

    fig.add_annotation(
        x=x,
        y=y,
        text=dim,
        showarrow=False,
        font=dict(size=10, color="white"),
        textangle=tang,
        xanchor="center",
        yanchor="middle"
    )

# ---------------------------------------------------------
# 10b. Ecological dimension labels just OUTSIDE the eco dark ring
# ---------------------------------------------------------
r_eco_dim_label = R_ECO_OUT + 0.08   # slightly outside the dark green ring

for dim in eco_dims:
    first = eco_dim_first_idx[dim]
    count = eco_dim_counts[dim]
    center_index = first + count / 2.0
    ang = center_index * eco_span      # angle in degrees
    t_lab = np.radians(ang)

    x = r_eco_dim_label * np.cos(t_lab)
    y = r_eco_dim_label * np.sin(t_lab)

    # same rotation rule as social dimensions
    pos_angle = ((ang + 180) % 360) - 180

    if pos_angle >= 0:
        tang = 90 - pos_angle
    else:
        tang = -90 - pos_angle

    fig.add_annotation(
        x=x,
        y=y,
        text=dim,
        showarrow=False,
        font=dict(size=10, color="black"),
        textangle=tang,
        xanchor="center",
        yanchor="middle"
    )

# ---------------------------------------------------------
# 11. SOCIAL FOUNDATION & ECOLOGICAL CEILING labels (curved)
# ---------------------------------------------------------
ang_top = 90  # centre of arc at the top (degrees)

r_sf_label  = 0.5 * (R_SF_IN  + R_SF_OUT)
r_eco_label = 0.5 * (R_ECO_IN + R_ECO_OUT)

add_arc_label(
    fig,
    "SOCIAL FOUNDATION",
    radius=r_sf_label,
    angle_center=ang_top,
    arc_span_deg=70,
    size=12
)

add_arc_label(
    fig,
    "ECOLOGICAL CEILING",
    radius=r_eco_label,
    angle_center=ang_top,
    arc_span_deg=70,
    size=12
)

# ---------------------------------------------------------
# 12. Ecological wedges (outer overshoot bars, per INDICATOR)
# ---------------------------------------------------------
for k, (dim, ind) in enumerate(eco_indicator_pairs):
    row = eco[(eco["dimension"] == dim) & (eco["indicator"] == ind)].iloc[0]
    frac = row["overshoot_frac"]
    raw  = row["overshoot_raw"]
    pct  = row["shortfallOvershoot_pct"]

    start_ang = k * eco_span
    end_ang   = (k + 1) * eco_span

    color = get_eco_color(raw)
    hover = f"{ind}: {pct:.1f}%"
    add_outward_bar(fig, frac, start_ang, end_ang, hover, color)

# ---------------------------------------------------------
# 13. Layout
# ---------------------------------------------------------
fig.update_layout(
    width=750,
    height=750,
    margin=dict(l=20, r=20, t=20, b=20),
    plot_bgcolor="white",
    paper_bgcolor="white",
    hoverlabel=dict(bgcolor="white", font_color="black")
)

# how tight you want the default crop:
VIEW_R = R_ECO_OUT + 1   # you already tuned this; keep as you like

fig.update_xaxes(
    visible=False,
    range=[-VIEW_R, VIEW_R]
)
fig.update_yaxes(
    visible=False,
    range=[-VIEW_R, VIEW_R],
    scaleanchor="x",
    scaleratio=1
)

```
